## Anomaly Agent

```python
import os
import asyncio
import time
import re
from collections import defaultdict
import csv

from app.controllers.ObjectaxController import objectax_controller
from app.services.objectax.AnomalyAgent import AnomalyAgentService
# from app.services.objectax.BnEdAgent import BnedAgentService
from app.generative.manager import manager

# ✅ Change this to your folder path
image_folder = "ordinary_box"
# image_folder = "cardboard box"

# Accept multiple image extensions
valid_exts = (".jpg", ".jpeg", ".png")
all_images = [
    os.path.join(image_folder, f) 
    for f in os.listdir(image_folder) 
    if f.lower().endswith(valid_exts)
]

all_images.sort()  # keep consistent order

# Initialize the real LLM once
llm = manager._get_llm(name="gemini_model_flash2", temperature=0)
agent = AnomalyAgentService(llm=llm)

def extract_batch(filename: str):
    """
    Extract batch from filename.
    Example: '16_1.jpg' → '16', '17a.png' → '17'
    """
    base = os.path.splitext(os.path.basename(filename))[0]
    match = re.match(r"([0-9]+)", base)
    return match.group(1) if match else base

def group_images_by_batch(images):
    """Group image files by batch prefix."""
    batches = defaultdict(list)
    for img in images:
        batch = extract_batch(img)
        batches[batch].append(img)
    return batches

async def process_batch(batch_number, image_files):
    """Process all images for a single batch."""
    try:
        # Load all images in the batch
        real_images = []
        for img in image_files:
            with open(img, "rb") as f:
                real_images.append(f.read())

        print(f"\nProcessing batch: {batch_number} with {len(real_images)} images")

        # Crop and upload
        state = {"images": real_images, "po_number": "PO1234"}
        urls = await objectax_controller.crop_and_upload_node(state=state)

        if not urls or "url" not in urls or not urls["url"]:
            raise RuntimeError(f"crop_and_upload_node returned invalid: {urls}")
        
        # Run the agent once for the whole batch
        url = urls['url']

        agent_state = {"url": url}
        
        result = await agent(agent_state)

        # print(result)
        return {
            "batch": batch_number,
            "result": result,
            "anomaly_result": result['anomaly']['is_anomaly'],
            "anomaly_reason": result['anomaly']['anomaly_reason'],
            "anomaly_model": result['anomaly_usage']['model'],
            "anomaly_prompt_tokens": result['anomaly_usage']['prompt_tokens'],
            "anomaly_completion_tokens": result['anomaly_usage']['completion_tokens'],
            "anomaly_total_tokens": result['anomaly_usage']['total_tokens'],
            "anomaly_cost": result['anomaly_usage']['cost'],
            "anomaly_runtime_sec": round(result['anomaly_runtime'], 2),
        }

    except Exception as e:
        print(f"⚠️ Skipping batch {batch_number!r} due to error: {e}")

async def run_batches(selected_batches: list[str] = None, csv_path: str = "results_anomaly.csv"):
    """
    Run batches and save results to a CSV file.
    - If selected_batches is None → run all batches
    - If selected_batches is a list → only run those (e.g., ["16", "19", "20"])
    """
    results = []
    grouped = group_images_by_batch(all_images)

    with open(csv_path, "w", newline="", encoding="utf-8") as csvfile:
        writer = csv.DictWriter(
            csvfile,
            fieldnames=["batch", "result", "anomaly_result", "anomaly_reason", 
                        "anomaly_model", "anomaly_prompt_tokens", "anomaly_completion_tokens",
                        "anomaly_total_tokens", "anomaly_cost", "anomaly_runtime_sec"]
        )
        writer.writeheader()

        for batch_number, imgs in grouped.items():
            if selected_batches and batch_number not in selected_batches:
                continue  # skip non-selected batches

            output = await process_batch(batch_number, imgs)
            results.append(output)

            # Write immediately for safety
            row_copy = output.copy()
            row_copy["result"] = str(row_copy["result"])  # stringify for CSV
            writer.writerow(row_copy)
            csvfile.flush()

    return results
```
